### MYC Targets

```r
file.path <- paste0(getwd(), "/lab-signatures/resources/RN_MYC_TARGETS/")
file <- list.files(file.path, full.names = T)
file <- file[grepl(".csv", file)]
file
```

```r
gsnames <- paste0("RN_MYC_", c("CRUDE", "CRUDE_CONTROL", "CRUDE_PRETUMOUR", "CRUDE_TUMOUR", "PROMOTER", "PROMOTER_CONTROL", "PROMOTER_PRETUMOUR", "PROMOTER_TUMOUR"))
genelist <- list()
for(i in seq_along(file)){
    genelist[[i]] <- read.csv(file[i], row.names = 1) %>%
        .$Gene.Name %>%
        unique(.)
    names(genelist)[i] <- gsnames[i]}
rn_myc_targets <- convert_list_to_df(genelist)
write_gmt(rn_myc_targets, save_dir = "lab-signatures-mouse/", save_name = "RN_MYC_TARGETS.gmt")
```
