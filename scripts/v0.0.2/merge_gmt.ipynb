```r
dir <- c("human", "mouse")
for(i in seq_along(dir)){

    gmt.ex <- list.files(paste0(dir[i],"-external/gmt/"), full.names = T)
    gmt.ex.list <- list()
    for(j in seq_along(gmt.ex)){
        ex.list <- read_gmt(gmt.ex[j])
        ex.list <- df_to_list(ex.list)
        gmt.ex.list <- c(gmt.ex.list, ex.list)}

    gmt.in <- list.files(paste0(dir[i],"-internal/gmt/"), full.names = T)
    gmt.in.list <- list()
    for(j in seq_along(gmt.in)){
        in.list <- read_gmt(gmt.in[j])
        in.list <- df_to_list(in.list)
        gmt.in.list <- c(gmt.in.list, in.list)}

    gmt.list <- c(gmt.ex.list, gmt.in.list)
    write_gmt(gmt.list, file = paste0("releases/v0.0.1/", dir[i], ".gmt"))}
```
