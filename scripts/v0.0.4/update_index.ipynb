```r
library(tidyverse)
library(readxl)
library(writexl)
library(strpip)
setwd("/Users/hungm/The Francis Crick Dropbox/CaladoD/Matthew/git-repos/projects/Calado-SignatureDB")
```

```r
external <- read_gmt("releases/v0.0.4/external.gmt") %>% df_to_list()
internal <- read_gmt("releases/v0.0.4/internal.gmt") %>% df_to_list()
combined <- c(external, internal) %>% list_to_df()
```

```r
# gmt files
gmt_ex <- list.files("external/gmt/", full.names = F) %>% gsub(".gmt", "", .)
gmt_in <- list.files("internal/gmt/", full.names = F) %>% gsub(".gmt", "", .)

# nsets
nset_ex <- c()
nset_in <- c()
for(i in seq_along(gmt_ex)){
    nset_ex[i] <- read_gmt(paste0("external/gmt/", gmt_ex[i], ".gmt")) %>% ncol()
}
for(i in seq_along(gmt_in)){
    nset_in[i] <- read_gmt(paste0("internal/gmt/", gmt_in[i], ".gmt")) %>% ncol()
}
```

```r
index <- data.frame(GMT = c(gmt_ex, gmt_in), N_SETS = c(nset_ex, nset_in), SOURCE = c(rep("EXTERNAL", length(gmt_ex)), rep("INTERNAL", length(gmt_in))), DOI = rep(NA, length(c(gmt_ex, gmt_in))), NOTES = rep(NA, length(c(gmt_ex, gmt_in))))
write_xlsx(index, "releases/v0.0.4/index.xlsx")
```
