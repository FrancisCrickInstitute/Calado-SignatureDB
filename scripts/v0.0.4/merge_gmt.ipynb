```r
read_gmt <- function(gmt) {
    stopifnot(file.exists(gmt))
    lines <- readLines(gmt)
    split_lines <- strsplit(lines, "\t")
    max_cols <- max(sapply(split_lines, length))
    padded <- lapply(split_lines, function(x) {
        length(x) <- max_cols
        x
    })
    mat <- do.call(rbind, padded)
    gene_names <- mat[, 1]
    mat_genes <- mat[, -c(1, 2), drop = FALSE]
    df <- as.data.frame(t(mat_genes), stringsAsFactors = FALSE)
    colnames(df) <- gene_names
    rownames(df) <- NULL
    return(df)
}
```

```r
external.gmt <- list.files("external/gmt/", full.names = T)
external.list <- c()
for(i in seq_along(external.gmt)){
    sig <- read_gmt(external.gmt[i])
    sig <- df_to_list(sig)
    external.list <- c(external.list, sig)}
write_gmt(external.list, file = paste0("releases/v0.0.4/external.gmt"))
```
